```python
df["x"] = np.random.rand(len(df.index)) * 100000 * np.random.rand(len(df.index))
df["y"] = np.random.rand(len(df.index)) * 100000
```

```python
np.random.rand(5)
```

```python
plt.rcParams.update(mpl.rcParamsDefault)
plt.figure(figsize=(10,5))

for i in list(df["value"].unique()):
    marker = np.random.choice([".",",","o","v","^","<",">","1","2","3","4","8","s","p","P",
                               "*","h","H","+","x","X","d","D","|","_"])

plt.xticks([])
plt.yticks([])

plt.xlim(1000,80000)
plt.ylim(1000,80000)
    
plt.scatter(df["x"], df["y"], marker=marker, c=df["value"], s=df["y"]/1e3, alpha=0.5)
plt.show();
```

```python
dfs = df.sample(1000)

style = np.random.choice(["dark_background"])#, "fivethirtyeight", "ggplot"])
plt.style.use(style)
plt.figure(figsize=(15,10))

plt.xticks([])
plt.yticks([])

plt.xlim(10000,20000)
plt.ylim(10000,20000)

for i in range(3):
    marker = np.random.choice([".",",","o","v","^","<",">","1","2","3","4","8","s","p","P",
                               "*","h","H","+","x","X","d","D","|","_"])
    plt.scatter(dfs["x"], dfs["y"], marker=marker, c=dfs["value"], s=dfs["value"]/10, alpha=0.5)

plt.show();
```
